## 2.3 reg_pct_level

Let us keep the columns according to the assignment guidelines

````python
# List of columns to keep
columns_to_keep = [
    'report_school_year', 'aggregation_index', 'aggregation_type', 'nrc_code',
    'nrc_desc', 'county_code', 'county_name', 'nyc_ind', 'membership_desc',
    'subgroup_code', 'subgroup_name', 'enroll_cnt', 'grad_cnt', 'grad_pct',
    'reg_cnt', 'reg_pct', 'dropout_cnt', 'dropout_pct'
]

# Keep only the specified columns
school_df = school_df[columns_to_keep]
````

Now let us create the new variable accoding to the guidelines

1. **Handling Missing Values:**
   - First, we check for missing or invalid values in the 'reg_pct' column, represented by '-'. We replace these with NaN (Not a Number) to ensure accurate calculations.

2. **Converting to Numeric Format:**
   - The 'reg_pct' column likely contains percentage symbols ('%'). To perform numerical operations, we remove these symbols and convert the column to a float data type.

3. **Calculating Median Value:**
   - We compute the median value of 'reg_pct'. The median is the middle value in a dataset, separating it into two equal halves. It serves as a reference point for creating categorical groups.

4. **Creating Categorical Indicator Variable:**
   - Now, we create a new categorical variable, 'reg_pct_level', based on the 'reg_pct' values.
   - We define three categories: 'low', 'medium', and 'high', which represent different ranges of regents diplomas percentages.
   - The bin edges are set according to the calculated median, creating distinct categories based on whether the regents percentage is below 0.5 times the median, between 0.5 and 1.5 times the median, or above 1.5 times the median.

5. **Displaying the Result:**
   - Finally, we display the updated DataFrame, showcasing the 'reg_pct' column and the newly created 'reg_pct_level' column for the first few rows.
  
This process transforms the original percentage data into a categorical variable, making it more manageable for analysis and interpretation. The 'reg_pct_level' variable helps classify school districts or student subgroupings into meaningful groups based on their regents diplomas percentages.

````python

# Replace '-' with NaN in 'reg_pct' column
school_df['reg_pct'] = school_df['reg_pct'].replace('-', np.nan)

# Remove percentage symbols and convert 'reg_pct' column to numeric
school_df['reg_pct'] = school_df['reg_pct'].replace('%', '', regex=True).astype('float')

# Calculate the median percentage of regents diplomas awarded
median_reg_pct = school_df['reg_pct'].median()

# Create the categorical indicator variable
school_df['reg_pct_level'] = pd.cut(
    school_df['reg_pct'],
    bins=[-float('inf'), 0.5 * median_reg_pct, 1.5 * median_reg_pct, float('inf')],
    labels=['low', 'medium', 'high'],
    include_lowest=True
)

# Display the updated DataFrame
print(school_df[['reg_pct', 'reg_pct_level']].head())

````

#### 7.5 Feature Scaling with StandardScaler



**Explanation:**

Feature scaling is a crucial preprocessing step in machine learning, especially when using algorithms that are sensitive to the scale of input features. StandardScaler is a popular method that standardizes the features by removing the mean and scaling to unit variance.

- **Standardization:** It transforms the data to have a mean of 0 and a standard deviation of 1. This ensures that all features are on a similar scale, preventing some features from dominating others during the model training process.

````python
# Standardize the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

**What was done and explanation:**

1. An instance of the `StandardScaler` is created, which will be used to scale the features.

2. The `fit_transform` method is applied to the training data (`X_train`), which computes the mean and standard deviation necessary for standardization and then transforms the training features.

3. The `transform` method is used on the test data (`X_test`). It uses the mean and standard deviation computed from the training data to ensure consistent scaling between training and test sets.

In summary, this step standardizes the features, making them comparable and preventing issues that may arise when using algorithms sensitive to feature scales.

#### 7.6 Applying Principal Component Analysis (PCA) for Dimensionality Reduction



**Explanation:**

In this step, we are utilizing Principal Component Analysis (PCA) to reduce the dimensionality of our feature matrix (`X`). Dimensionality reduction is beneficial for various reasons, including mitigating the curse of dimensionality, speeding up model training, and simplifying the model.

- **PCA:** PCA is a technique used to transform high-dimensional data into a lower-dimensional representation while retaining the most crucial information. It achieves this by identifying the principal components, which are linear combinations of the original features.

- **n_components=0.95:** We are specifying that PCA should retain 95% of the variance in the original data. By retaining a high percentage of the variance, we aim to preserve the essential information while reducing the number of features.

````python
# Apply PCA to reduce dimensionality
pca = PCA(n_components=0.95)  # Retain 95% of the variance
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
````

**What was done and explanation:**

1. An instance of the PCA class is created with the parameter `n_components=0.95`, indicating that the transformed data should retain 95% of the variance from the original features.

2. The `fit_transform` method is employed on the training set (`X_train_scaled`), which fits the PCA model to the data and transforms the feature matrix into a lower-dimensional representation. The resulting transformed feature matrix is assigned to `X_train_pca`.

3. The `transform` method is applied to the test set (`X_test_scaled`), using the parameters learned from the training set. The transformed test set is stored in the variable `X_test_pca`.

In essence, this step reduces the dimensionality of our feature matrix while retaining the majority of the original information, facilitating more efficient and effective model training.

#### 7.7 ANOVA F-Statistic Feature Selection



**Explanation:**

In this section, we employ the ANOVA F-statistic method for feature selection. ANOVA (Analysis of Variance) assesses the variation between groups and within groups in a dataset. The F-statistic is a ratio of two variances, and in the context of feature selection, it helps identify features that significantly impact the target variable.

- **Objective:** To choose the top 10 features that have the most significant impact on the target variable 'reg_pct_level.'

````python
# Apply feature selection using ANOVA F-statistic
k_best = SelectKBest(f_classif, k=10)  # Select top 10 features
X_train_k_best = k_best.fit_transform(X_train_scaled, y_train)
X_test_k_best = k_best.transform(X_test_scaled)
````

**What was done and explanation:**

1. **Initialization:** An instance of the `SelectKBest` class is created, specifying the ANOVA F-statistic (`f_classif`) as the scoring method and selecting the top 10 features (`k=10`).

2. **Fit and Transform:** The `fit_transform` method is applied to the scaled training set (`X_train_scaled`) and the target variable (`y_train`). This operation identifies and retains the top 10 features based on their F-statistic scores, creating a new feature matrix `X_train_k_best`.

3. **Transform Test Set:** The same feature selection is applied to the scaled test set (`X_test_scaled`) using the `transform` method, resulting in the creation of a new test set feature matrix `X_test_k_best`.

In essence, this step is geared towards narrowing down the feature set to the most influential ones, as determined by the ANOVA F-statistic. The selected features are deemed to contribute significantly to predicting the target variable 'reg_pct_level' and are utilized in subsequent modeling steps.

#### 7.8 Recursive Feature Elimination (RFE) with Random Forest



**Explanation:**

In this section, we employ Recursive Feature Elimination (RFE) in combination with a Random Forest classifier. RFE is a feature selection technique that recursively removes the least important features, and Random Forest is used as the estimator to evaluate feature importances.

- **Random Forest Classifier (`rf_classifier`):** Random Forest is an ensemble learning method that constructs multiple decision trees during training and outputs the mode of the classes (classification) of the individual trees.

- **RFECV (`rfe`):** Recursive Feature Elimination with Cross-Validation is a method that assesses the importance of each feature through cross-validation. It iteratively removes features, evaluates model performance, and retains the subset of features that yields the best performance.

- **Feature Transformation (`X_train_rfe` and `X_test_rfe`):** The dataset is transformed to include only the selected features deemed important by the RFE process. This reduces the dimensionality of the feature space.

````python
# Apply Recursive Feature Elimination (RFE) with Random Forest
rf_classifier = RandomForestClassifier()
rfe = RFECV(estimator=rf_classifier, step=1, cv=5)
X_train_rfe = rfe.fit_transform(X_train_scaled, y_train)
X_test_rfe = rfe.transform(X_test_scaled)
````

**What was done and explanation:**

1. A Random Forest Classifier (`rf_classifier`) is instantiated. This classifier will be used to evaluate the importance of each feature during the RFE process.

2. RFECV is configured with the Random Forest classifier, setting the step size for feature elimination to 1 and utilizing 5-fold cross-validation (`cv=5`). Cross-validation ensures robust evaluation of feature importance.

3. The `fit_transform` method is applied to the scaled training data (`X_train_scaled`, `y_train`). This initiates the RFE process, identifying and retaining the most important features based on Random Forest evaluation.

4. The `transform` method is then used to apply the same feature selection to the scaled testing data (`X_test_scaled`). This ensures consistency in feature dimensions between the training and testing datasets.

In summary, this section implements Recursive Feature Elimination with Random Forest, resulting in a reduced set of features that are expected to contribute significantly to the predictive performance of the model.
